#### Exercise 09. Is the average of "Valdemorillo" and "Galapagar" prices the same? (★★☆)

Print the both average prices and then write a conclusion about them

```python
import pandas as pd

ds = pd.read_csv('assets/real_estate.csv', sep=';')
ds

valdemorillo_prices = ds.loc[ds['level5'] == "Valdemorillo", 'price']
galapagar_prices = ds.loc[ds['level5'] == "Galapagar", 'price']

average_valdemorillo_price = valdemorillo_prices.mean()
average_galapagar_price = galapagar_prices.mean()

print("Average price in Valdemorillo:", average_valdemorillo_price, "USD.")
print("Average price in Galapagar:", average_galapagar_price, "UDS.")

if average_valdemorillo_price == average_galapagar_price:
    print("The average prices in Valdemorillo and Galapagar are the same.")
else:
    print("The average prices in Valdemorillo and Galapagar are different.")
```

The average prices are different. The averge price of Valdemorillo is more expensive than the avergae price of Galapagar by a total of 3797.0907225 USD. 

#### Exercise 10. Is the average of "Valdemorillo" and "Galapagar" price per square meter (price/m2) the same? (★★☆)

Print the both average prices and then write a conclusion about

Hint: Create a new column called `pps` (price per square) and then analyse the values

```python
valdemorillo_data = ds.loc[ds['level5'] == "Valdemorillo", ['price', 'surface']]
galapagar_data = ds.loc[ds['level5'] == "Galapagar", ['price', 'surface']]

valdemorillo_data['pps'] = valdemorillo_data['price'] / valdemorillo_data['surface']
galapagar_data['pps'] = galapagar_data['price'] / galapagar_data['surface']

average_valdemorillo_pps = valdemorillo_data['pps'].mean()
average_galapagar_pps = galapagar_data['pps'].mean()

print("Average price per square meter in Valdemorillo:", average_valdemorillo_pps)
print("Average price per square meter in Galapagar:", average_galapagar_pps)

if average_valdemorillo_pps == average_galapagar_pps:
    print("The average price per square meter in Valdemorillo and Galapagar is the same.")
else:
    print("The average price per square meter in Valdemorillo and Galapagar is different.")
```

#### Exercise 16. Calculate the sample mean and variance of the variables: price, rooms, surface area and bathrooms (you must use the subset obtained in the question 14) (★★★)

Print both values for each variable

```python
import pandas as pd

ds = pd.read_csv('assets/real_estate.csv', sep=';')
ds

south_belt_populations = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]

south_belt_subset = ds[ds['level5'].isin(south_belt_populations)]

mean_price = south_belt_subset['price'].mean()
variance_price = south_belt_subset['price'].var()

mean_rooms = south_belt_subset['rooms'].mean()
variance_rooms = south_belt_subset['rooms'].var()

mean_surface_area = south_belt_subset['surface'].mean()
variance_surface_area = south_belt_subset['surface'].var()

mean_bathrooms = south_belt_subset['bathrooms'].mean()
variance_bathrooms = south_belt_subset['bathrooms'].var()

print("Sample Mean and Variance for Price:")
print("Mean Price:", mean_price)
print("Variance Price:", variance_price)

print("\nSample Mean and Variance for Rooms:")
print("Mean Rooms:", mean_rooms)
print("Variance Rooms:", variance_rooms)

print("\nSample Mean and Variance for Surface Area:")
print("Mean Surface Area:", mean_surface_area)
print("Variance Surface Area:", variance_surface_area)

print("\nSample Mean and Variance for Bathrooms:")
print("Mean Bathrooms:", mean_bathrooms)
print("Variance Bathrooms:", variance_bathrooms)
```

#### Exercise 18. Normalize the variable of prices for each population and plot the 4 histograms in the same plot (you must use the subset obtained in the question 14) (★★★)

```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(figsize=(10, 6))

populations = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]

normalized_prices = []

for population in populations:
    population_data = south_belt_subset[south_belt_subset['level5'] == population]
    prices = population_data['price']
    
    min_price, max_price = prices.min(), prices.max()
    normalized_price = (prices - min_price) / (max_price - min_price)
    
    normalized_prices.append(normalized_price)

plt.hist(
    normalized_prices,
    bins=20,
    alpha=0.7,
    label=populations
)

plt.title("Normalized Prices Histogram for Different Populations")
plt.xlabel("Normalized Price")
plt.ylabel("Frequency")
plt.legend(title="Population")

plt.show()
```
